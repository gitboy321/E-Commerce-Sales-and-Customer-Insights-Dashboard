```python
import csv
```

```python
with open('Orders_dirty.csv', 'r') as file:
    reader = csv.DictReader(file)
    data = list(reader)
```

```python
print(data[:5])
```

```python
missing_email = 0
missing_city = 0

for row in data:
    if row['Email'] == '':
        missing_email += 1
    if row['City'] == '':
        missing_city += 1

print("Missing Email:", missing_email)
print("Missing City:", missing_city)
```

```python
import re

pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'

for row in data:
    email = row['Email']
    if email != '' and not re.match(pattern, email):
        print("Invalid Email:", email)
```

```python
revenue = {}

for row in data:
    try:
        category = row['Category']
        qty = float(row['Quantity'])
        price = float(row['UnitPrice'])

        total = qty * price

        if category in revenue:
            revenue[category] += total
        else:
            revenue[category] = total
    except:
        pass

print(revenue)
```

```python
from collections import Counter

products = [row['Product'] for row in data]

count = Counter(products)

print(count.most_common(5))
```

```python
from datetime import datetime

month_count = {}

for row in data:
    try:
        date = datetime.strptime(row['OrderDate'], "%d-%m-%Y")
        month = date.month

        if month in month_count:
            month_count[month] += 1
        else:
            month_count[month] = 1
    except:
        pass

print(month_count)
```
